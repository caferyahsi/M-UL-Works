```python
for i in range(len(df["AGE"])):
    if df.loc[i,"AGE"] < 18:
        df.loc[i,"AGE"]="0_18"
    elif df.loc[i,"AGE"]<23:
        df.loc[i,"AGE"]="19_23"
    elif df.loc[i,"AGE"]<30:
        df.loc[i,"AGE"]="24_30"
    elif df.loc[i,"AGE"]<40:
        df.loc[i,"AGE"]="31_40"
    elif df.loc[i,"AGE"]<70:
        df.loc[i,"AGE"]="41_70"
df["AGE"]            
```
